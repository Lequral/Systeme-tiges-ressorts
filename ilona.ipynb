```python
import sympy as sp
from sympy.physics.mechanics import dynamicsymbols, Point, ReferenceFrame
#BARRE 2
#paramètres
r_21 , r_22, r_3, l, m2, g, m3, theta_20, gamma_10 = sp.symbols('r_21 r__22 r_3 l m2 g m3 theta_20 gamma_10', real = True, positive = True, nonzero = True)

# paramètres de position
theta_2, gamma_1 = sp.symbols('theta_2 gamma_1', real = True)


# efforts 
F_2, T_2, X02, Y02, P2, P3, X03, Y03 , T_3, F_3= sp.symbols('F_2 T_2 X02 Y02 P2 P3 X03 Y03 T_3 F_3')

# raideur du ressort
k2 = sp.symbols('k2', real = True, positive = True, nonzero = True)
k3 = sp.symbols('k3', real = True, positive = True, nonzero = True)

#repères
R0 = ReferenceFrame('R_0') 
R2 = ReferenceFrame('R_2')
R3 = ReferenceFrame('R_3')

#position des repères
R2.orient(R0,'Axis',[theta_2, R0.z])
R3.orient(R0,'Axis',[gamma_1, R0.z])

# Points caractéristiques 

A2 = Point('A2')
O2 = Point('O2')
B2 = Point('B2')
C = Point('C')
D = Point('D')
G2 = Point('G_2')
G3 = Point('G_3')

# position associées 
O2.set_vel(R0, 0) 
O2.set_vel(R2, 0) 
O2.set_vel(R3, 0) 

A2.set_pos(O2, -r_21 * R2.x )
A2.set_vel(R2, 0.0) # A est un point de R2 donc vitesse nulle par rapport à 2

#
B2.set_pos(O2, r_22 * R2.x ) 
B2.set_vel(R2, 0.0) 

#
C.set_pos(O2, r_22 * R2.x + l * R0.y)
C.set_vel(R3, 0.0) 

#
D.set_pos(O2, r_22 * R2.x + l * R0.y + r_3 * R3.x)
D.set_vel(R3, 0.0) 
D.set_vel(R0, 0.0) 

#
G2.set_pos(O2, 0)
G2.set_vel(R2, 0.0) 

G3.set_pos(D,(r_3/2)* R3.x )
G3.set_vel(R3, 0.0) 


#BAME sur la barre 1

# Liaison 0/2
R_02 = X02 * R0.x + Y02 * R0.y # Pivot et analyse plane

# actions extérieure 
R_P2 = -m2*g * R0.y # en G2
M_P2 = G2.pos_from(O2).cross(R_P2)

T_2 = T_2 * R2.y 
M_T_2 = B2.pos_from(O2).cross(T_2)

F_2 = F_2 * R2.y
M_F_2 = A2.pos_from(O2).cross(F_2)

MO2 = k2 *(theta_2 - theta_20)*R0.z

#PFS 

R_x = R_02.dot(R0.x) + R_P2.dot(R0.x) + T_2.dot(R0.x)+ F_2.dot(R0.x)
R_y = R_02.dot(R0.y) + R_P2.dot(R0.y) + T_2.dot(R0.y)+F_2.dot(R0.y)
Eq_Rx_1 = sp.Eq(R_x, 0)
Eq_Ry_1 = sp.Eq(R_y, 0)

O2A2 = A2.pos_from(O2) #donne A202
M_T_2 = B2.pos_from(O2).cross(T_2)
M_T_2 = B2.pos_from(O2).cross(T_2)
M = MO2.dot(R0.z) + M_P2.dot(R0.z)+ M_T_2.dot(R0.z) + M_F_2.dot(R0.z)

Eq_M_O2 = sp.Eq(M, 0)

print('Théorème de la résultante à S2 isolé')
display(Eq_Rx_1, Eq_Ry_1, Eq_M_O2)

#BAME barre 3

# Liaison 0/3
R_03 = X03 * R0.x + Y03 * R0.y # Pivot et analyse plane

# actions extérieure 
R_P3 = -m3*g * R0.y # en G3 = D

T_3 = T_3 * R3.y 
M_T_3 = C.pos_from(D).cross(T_3)

F_3 = F_3 * R3.y
M_F_3 = C.pos_from(D).cross(F_3)

MO3 = k3 *(gamma_1 - gamma_10)*R0.z

#PFS 

R_x_3 = R_03.dot(R0.x) + R_P3.dot(R0.x) + T_3.dot(R0.x)+ F_3.dot(R0.x)
R_y_3 = R_03.dot(R0.y) + R_P3.dot(R0.y) + T_3.dot(R0.y)+F_3.dot(R0.y)
Eq_Rx_3 = sp.Eq(R_x, 0)
Eq_Ry_3 = sp.Eq(R_y, 0)

M3 = MO3.dot(R0.z) + M_T_3.dot(R0.z) + M_F_3.dot(R0.z)

Eq_M_O3 = sp.Eq(M3, 0)

print('Théorème de la résultante à S3 isolé')
display(Eq_Rx_3, Eq_Ry_3, Eq_M_O3)


```

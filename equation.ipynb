```python
import calcul_analytique
import sympy as sp
from random import randint

weights = [randint(-1000,1000)/100 for i in range(3)]
print("poids",weights)

solu = calcul_analytique.trouver_param_real(weights)

print("optimalité de la solution",solu.optimality)
assert solu.optimality < 9e-3

param_value = solu.x
print(solu.x)
param_value = [round(e, 3) for e in param_value] # les paramètres du problèmes n'ont pas une précision infinie

data = calcul_analytique.equations(True)
w_expr, param = data["weight_expression"], data["param"]
r_11, r_12, r_21, r_22, r_3, k_1, k_2, k_3, b = param


for ii in range(len(w_expr)):
    e = sp.simplify(w_expr[ii].subs(
        {param[i]: param_value[i] for i in range(len(param))}))
    print("poids réel",e,"poids théoriques",weights[ii])
    assert abs(e-weights[ii]) < 0.5
```
